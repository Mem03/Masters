```python
# Legendre polynomial for order n with x lags
def legendre_polynomial(x, n):
    """
    This function calculates the Legendre polynomial of order n for the input x.
    For degrees above 2, the function uses the recursive relation to calculate the polynomial.
    """
    if n == 0:
        return 1
    elif n == 1:
        return x
    else:
        return ((2 * n - 1) * x * legendre_polynomial(x, n - 1) - (n - 1) * legendre_polynomial(x, n - 2)) / n
```

```python
# Create matrix og legendre polynomials for lags and degrees
def legendre_matrix(x_lags, legendre_degree = 3):
    """
    This function creates a matrix of Legendre polynomials for the input x_lags and legendre_degree.
    """
    # Equally spaced values between -1 and 1
    x_values = np.linspace(-1, 1, num=x_lags)
    # Loop through the lags
    legendre_matrix = [[legendre_polynomial(x, n) for n in range(legendre_degree)] for x in x_values]
    return legendre_matrix
```
